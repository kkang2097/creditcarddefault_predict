# Data Exploration

Let's take a look at the data first using df.head()...

Note that there are only numeric variables since a PCA transformation has already been done on the data.

```python
import pandas as pd

df = pd.read_csv('C:\\Users\\Elliot\Desktop\\data science datasets\\credit card fraud\\creditcard.csv')
df.head(n=10)
```

I doubt 'Time' would have any predictive capability or correlation to when a fraud would occur, but maybe credit card fraud is intervalistic (if that's a word). Let's check by putting 'Time' on the x-axis and 'class' on the y-axis for a scatterplot.

```python
import matplotlib.pyplot as plt
import numpy as np

#Define which observations are frauds
frauds = df.loc[df['Class'] == 1]
non_frauds = df.loc[df['Class'] == 0]

ax = frauds.plot.scatter(x='Time', y='Class', color='Red', label='Fraud')
non_frauds.plot.scatter(x='Time', y='Class', color='Green', label='Non-Fraud', ax=ax)
plt.show()
```

So the frauds don't seem to happen at a particular time, since the distribution isn't clustered heavily in one or more (strictly concentrated) areas. Next, let's see if Amount of Transaction indicates anything about how suspicious a transaction might be.

```python
ax = frauds.plot.scatter(x='Amount', y='Class', color='Orange', label='Fraud')
non_frauds.plot.scatter(x='Amount', y='Class', color='Blue', label='Non-Fraud', ax=ax)
plt.show()
```

High transaction amounts don't necessarily mean it is a fraudulent transaction. This could be also because banks scrutinize higher credit card transactions since the misclassification of high balance transactions can be incredibly costly. In addition, high amounts ($5000 USD+) are not even possible on most credit cards due to credit limit.

This concludes the data exploration, but more would have been required had this dataset not been cleaned/transformed with PCA prior to releasing the dataset on Kaggle. https://www.kaggle.com/dalpozz/creditcardfraud
